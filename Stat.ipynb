```python
#imports
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
#load data
df = pd.read_csv("Data.csv")
dt = df["DT"]
rf = df["RF"]
```

```python
#Mean and std
print("\nDT:")
print(f"Mean: {np.mean(dt):.2f}")
print(f"Std: {np.std(dt, ddof=1):.2f}")

print("\nRF:")
print(f"Mean: {np.mean(rf):.2f}")
print(f"Std: {np.std(rf, ddof=1):.2f}")
```

```python
# Confidensinterval (95%)

n_dt = len(dt)
n_rf = len(rf)

# DT
mean_dt = np.mean(dt)
se_dt = stats.sem(dt)
ci_dt = stats.t.interval(0.95, df=n_dt-1, loc=mean_dt, scale=se_dt)

# RF
mean_rf = np.mean(rf)
se_rf = stats.sem(rf)
ci_rf = stats.t.interval(0.95, df=n_rf-1, loc=mean_rf, scale=se_rf)

print("\n95% Confidence Intervals:")

print("\nDT:")
print(f"CI: ({ci_dt[0]:.2f}, {ci_dt[1]:.2f})")

print("\nRF:")
print(f"CI: ({ci_rf[0]:.2f}, {ci_rf[1]:.2f})")
```

```python
#Paired t-test
alpha = 0.05
t_stat, p_value = stats.ttest_rel(rf, dt)


print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.6f}")


```
